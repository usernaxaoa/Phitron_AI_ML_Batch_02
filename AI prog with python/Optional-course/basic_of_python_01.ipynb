```python
# Python inbuilt function

import math

print(math.ceil(3.4))
print(math.floor(34.34))
print(math.exp(3))  # math.exp(x) computes eˣ, where e ≈ 2.71828 (Euler’s number)
print(math.log(10))
print(math.log(10, 10))
print(math.sqrt(16))
print(abs(-34.34))
print(abs(34.34))
```

```python
# The ord and chr functions
# play with ascii values

print(ord('A')) # 65
print(ord('a')) # 97
print(chr(65)) # 'A'
print(chr(97)) # 'a'

print(round(1.5))
print("{:,}".format(1254216514322)) #1,254,216,514,322
# modern way
print(f"{233245346353:,}")

# "{:,}".format(x)	    add comma separators
# f"{x:,}"	            cleaner modern formatting
```
