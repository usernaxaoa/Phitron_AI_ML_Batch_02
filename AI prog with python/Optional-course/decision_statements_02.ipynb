```python
import math

radius = eval(input("Enter radius: "))
if radius > 0:
    print(f"ares: {math.pi * ((radius) ** 2):.3f}")
    print(f"circumferrence: {2 * math.pi * radius:.3f}")
else:
    print("Enter correct value")
```
