```python
digit = 12341234
new_digit = digit
sum = 0
while new_digit > 0:
    last_digit = new_digit % 10
    sum += last_digit
    new_digit = new_digit // 10

print(sum)
# print(last_digit)
# print(new_digit)
```

```python
digit = 12345
new_digit = digit
sum = 0
while new_digit > 0:
    last_digit = new_digit % 10
    sum = sum * 10 + last_digit
    new_digit = new_digit // 10

print(sum)
# print(last_digit)
# print(new_digit)
```

```python
fact = 1
num = 5
if num < 0:
    print('Enter a valid positive number')
else:
    while num > 0:
        fact  *= num
        num -= 1
    print(fact)
```
