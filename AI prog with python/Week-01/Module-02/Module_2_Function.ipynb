# Different type of argument


```python
def adder(a, b, c):
    print(f"{a},{b},{c}")


adder(10, 30, 20)
```

```python
def adder(*args):
    for i in args:
        print(i)


adder(10, 30, 20, 40)
```

```python
def adder(a=10, b=30, c=10):
    print(f"{a},{b},{c}")


adder()
```

# generator

```python
lst = [x for x in range(500)]  # dataset


def data_loader(chunk_size, lst):
    for i in range(0, len(lst), chunk_size):
        yield lst[i : i + chunk_size]


x = data_loader(5, lst)

print(next(x))
print(next(x))
```
