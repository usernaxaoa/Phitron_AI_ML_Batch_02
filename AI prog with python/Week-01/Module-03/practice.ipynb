```python
# problem 1

file = open("./sample_data/emails.txt", "r")
content = file.readlines()

lower_case_email = list(map(lambda x : x.lower(), content))
remove_duplicates_email = list(set(lower_case_email))
sorted_emails = sorted(remove_duplicates_email)

uniq_email = open("./sample_data/unique_emails.txt", "w")
uniq_content = uniq_email.writelines(sorted_emails)
```

```python
# problem 2

with open("./sample_data/names.txt", "r") as f:
  content = f.readlines()
  remove_space = list(map(lambda x : x.strip(), content))
  capitalized = list(map(lambda x : x.capitalize(), remove_space))
  remove_duplicate = list(set(capitalized))
  sorted_output = sorted(remove_duplicate)
  print(content)
  print(remove_space)
  print(capitalized)
  print(remove_duplicate)
  print(sorted_output)
```

```python
# problem 3

def custom_list_creation(index_number):
  cust_list = []
  try:
    for i in range(5):
      usr_inp = int(input("Enter a value: "))
      cust_list.append(usr_inp)

    if index_number >= len(cust_list):
      return ("IndexError")
    else:
       return (cust_list[index_number])

  except Exception as e:
    return (e)

index_number = int(input("Enter a index number between 0-4: "))
custom_list_creation(index_number)
```
