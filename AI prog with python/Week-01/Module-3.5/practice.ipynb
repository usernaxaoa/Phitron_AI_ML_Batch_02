```python
# problem 1

def create_file(filename, content):
  with open(f"./sample_data/{filename}", "w") as f:
    f.write(content)

filename = input("Enter the file name: ")
content = input("Enter the content: ")

create_file(filename, content)
```

```python
# problem 2

def read_lines(filename):
  try:
    f = open("./sample_data/data.txt", "r")
    c = f.readlines()
    count = 1
    for word in c:
      print(f'{count}: {word}', end="")
      count += 1
  except Exception as e:
    print(e)
  finally:
    f.close()

read_lines("data.txt")
```

```python
# Problem 3

def add_log(message):
  try:
    with open("./sample_data/log.txt", "w") as f:
      f.write(message)
  except Exception as e:
    pass


message = input("Enter a message: ")
add_log(message)
```

```python
# problem 4

def read_from_position(filename, position):
  try:
    f = open(f"./sample_data/{filename}", "r")
    f.seek(position)
    content = f.read()
    print(content)
  except Exception as e:
    print(e)
  finally:
    f.close()

filename = "file1.txt"
position = 7
read_from_position(filename, position)
```

```python
# problem 5

def safe_read(filename):
  try:
    with open(f"./sample_data/{filename}", "r") as f:
      c = f.read()
  except Exception as e:
      print(e)
  else:
    print(c)
  finally:
    print("Operation complete")

safe_read("missing.txt")
safe_read("hello.txt")
```
