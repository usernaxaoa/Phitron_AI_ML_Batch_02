```python
# parent class

class Phone:
    category = "electronics"

    # constructerr
    def __init__(self, model, battery, camera, battery_per):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_per = battery_per

    # methods
    def charge(self, hour):
        print(f"charge completed by {hour} hour")
    
    def capture(self, photo):
        if self.battery_per <= 0:
            print(f"no charge")
        else:
            self.battery_per -= photo
            print(f"photo capured in {self.model}")

# child class (derived class)
class SmartPhone(Phone):
    def __init__(self, model, battery, camera, battey_per, processor):
        super().__init__(model, battery, camera, battey_per)
        self.processor = processor
    
    # modified method
    def charge(self, hour):
        print(f"fast charging in progress", end=" | ")
        super().charge(hour)


class Cooling_Mechanism:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method

    def cooling_on(self):
        return "Cooling system activated"

class SmartPhone_CoolingMode(SmartPhone, Cooling_Mechanism):
    def __init__(self, model, battery, camera, battey_per, processor, cooling_method):
        SmartPhone.__init__(self, model, battery, camera, battey_per, processor)
        Cooling_Mechanism.__init__(self, cooling_method)

# method overloading
class Camera:
    def zoom(self, level=1):
        print(f"zoom level {level}")

z = Camera()
z.zoom()



# pro_cooling = SmartPhone_CoolingMode("y", 5000, 100, 100, "sd", "nitrogen")
# print(pro_cooling.cooling_method)
# print(pro_cooling.processor)
# print(pro_cooling.cooling_method)
# print(pro_cooling.cooling_on())
# print(pro_cooling.charge(100))

# pro = SmartPhone("x", 5000, 100, 5, "snapdragon")
# pro.capture(4)
# # pro.charge(1)

# apple = Phone('iphone17', 3500, 35, 90)
# motorolla = Phone('moto17', 4500, 20, 20)
# xiaomi = Phone('redim17', 6500, 100, 100)

# motorolla.charge(20)
# motorolla.capture(50)
# print(xiaomi.battery_percentage)
```

```python
class Camera:
    def __init__(self, name) -> None:
        self.name = name
    # parent method
    def captured(self):
        print("a photo is captured")

class Smart_Phone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation

    # method overwritting
    def captured(self):
        print(f"Smart phone")

class Dslr(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation

    # method overwritting
    def captured(self):
        print(f"Dslr")

class Drone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation

    # method overwritting
    def captured(self):
        print(f"Drone")
    


phone = Smart_Phone("Phone", 30)
dslr = Dslr("DSLR", 200)
drone = Drone("Drone", 150)

phone.captured()
dslr.captured()
drone.captured()
```

```python
devices = ['phone', 'dslr', 'drrone']
def captured_by(device):
    print(f"captured by {device}")
for device in devices:
    captured_by(device)
```

```python
class Mobile:
    def __init__(self, name, model, imei) -> None:
        self.__name = name
        self.__model = model
        self.__imei = imei

    def charge(self):
        print(f"phone is charging")
    
    # getter methods
    def name_getter(self):
        return self.__name
    def model_getter(self):
        return self.__model
    def imei_getter(self):
        return self.__imei
    
    # setter methods
    def name_setter(self, name):
        self.__name = name
    def imei_setter(self, name):
        self.__imei = name


iphone = Mobile("Phone", "17", "1xkaf1")
print(iphone.imei_getter())
iphone.imei_setter('changed imei')
print(iphone.imei_getter())
```

```python
from abc import ABC, abstractmethod

class Development(ABC):
    def database(self):
        print('connected to database')

    @abstractmethod
    def security(self):
        print("mobile security")
```

```python
class MobileApp(Development):
    def mobile_login(self):
        print("login into mobile")
    
    def security(self):
        print("mobile security")
```

```python
mob = MobileApp()
mob.security()
mob.database()
mob.mobile_login()
```
