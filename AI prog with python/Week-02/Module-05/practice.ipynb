```python
class Phone:
    def __init__(self, model, battery=100) -> None:
        self.model = model 
        self.battery = battery 

class SmartPhone(Phone):
    def __init__(self, model, os, battery=100) -> None:
        super().__init__(model, battery)
        self.os = os

class GamingPhone(SmartPhone):
    def __init__(self, model, os, cooling_system, battery=100) -> None:
        Phone.__init__(self, model, battery=100)
        SmartPhone.__init__(self, model, os, battery=100)
        self.cooling_system = cooling_system

    def start_game(self, name):
        print(f"playing {name} on {self.model}")

test = GamingPhone('iphone 17', 'android', 'vapor', 5000)
print(test.cooling_system)
print(test.battery)
print(test.model)
test.start_game("Nafiz")
```

```python
class MobilePhone:
    def __init__(self, model, battery=100) -> None:
        self.model = model
        self.__battery = battery

    def battery_getter(self):
        print(self.__battery)
    
    def use(self, used_battery):
        print(f"Starting battery percentage: {self.battery_getter()}")
        if used_battery > self.__battery:
            print("Enter valid battery use percentage.")
        else:
            self.__battery -= used_battery
            print(f"after use {used_battery}% now current mobile battery: {self.__battery}")
    
    def charge(self, charged):
        if charged >= 100:
            self.__battery = 100
        else:
            self.__battery = self.__battery + charged
            print(f"after charge current battery: {self.__battery}")

    def get_battery(self):
        print(self.__battery)



mobile = MobilePhone('redmi 13c', 100)
mobile.use(70)
mobile.charge(50)
mobile.get_battery()
```
