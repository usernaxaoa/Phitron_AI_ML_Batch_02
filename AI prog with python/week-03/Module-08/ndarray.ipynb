```python
import numpy as np
```

```python
# 1d array

arr1 = np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.ndim)


```

```python
# 2d array

arr2 = np.array([[1,2,3], # row 1
                 [4,5,6] # row 2
                 ])

print(arr2)
print(arr2.ndim)
```

```python
# 3d array

arr3 = np.array([
    # 1st floor
    [[1,2,3],
    [4,5,6]],
    
    # 2nd floor
    [[1,2,3],
    [4,5,6]],
    # 3rd
    #  floor
    [[1,2,3],
    [4,5,6]],
])

print(arr3)
print(arr3.ndim)
```

```python
# array attributes

# dimension check
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
print()

# shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
print()

# data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)
print()

# size
print(arr1.size)
print(arr2.size)
print(arr3.size)
print()
```
