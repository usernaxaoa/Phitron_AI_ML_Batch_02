```python
# x = [10,20,30,40,50]
# y = [2,3,4,5]
# add = x + y
# add

x = np.array([1,4,9,16])
y = np.array([2,3,4,5])

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x // y)
print(x % y)
```

```python
# mathmathical function

# trigonomatry
# sin, cos, tan takes values in radian

sin_val = np.sin(x)
cos_val = np.cos(x)
tan_val = np.tan(x)

print(sin_val)
print(cos_val)
print(tan_val)

# deg to rad or rad to deg convestion

deg_con = np.rad2deg(x)
print(deg_con)

rad_con = np.deg2rad(deg_con)
print(rad_con)
```

```python
# log
base_10_log = np.log10(x)
# print(base_10_log)

base_2_diye_vag = np.log2(x)
# print(base_2_diye_vag)

sqrt = np.sqrt(x)
# print(sqrt)
```

```python
# different

jog = np.sum(x)
cum_jog = np.cumsum(x)

print(jog)
print(cum_jog)
```

```python
# logical functions

x = np.array([10, 8, 16, 100])
y = np.array([2, 8, 4, 500])

greater_than = x > y
equal = x == y
print(greater_than)
print(equal) 

# all
# to check sob trrue kina
# any 
# to check jekono ekta true kina

print(np.all(greater_than))
print(np.any(greater_than))
```
