```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2,1,3,5,3,1,4]
days = [1,2,3,4,5,6,7]

plt.plot(days, hours)
```

```python
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Trend of a student study time over a week")

plt.plot(days, hours)
```

```python
hours1 = [2,1,3,5,3,1,4]
hours2 = [3,5,1,2,0,2,3]

days = [1,2,3,4,5,6,7]

plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Trend of a student study time over a week")

plt.plot(days, hours1)
plt.plot(days, hours2)

```

```python
maindf = pd.read_csv('temperature.csv')

df = maindf.sample(5)
df2 = maindf.sample(5)

plt.xlabel("id")
plt.ylabel("temp")
plt.title("id and temp")

plt.plot(df['record_id'], df['AverageTemperatureFahr'], label="df")
plt.plot(df2['record_id'], df2['AverageTemperatureFahr'], label="df2")

plt.legend(loc='best')
```

```python
# adding some visualization

plt.plot(df['record_id'], df['AverageTemperatureFahr'], label="df", color='#ff00fb', linewidth=2, linestyle='dotted', marker='o', markersize='10')
plt.plot(df2['record_id'], df2['AverageTemperatureFahr'], label="df2", color='#0095ff', linewidth=2, linestyle='dotted', marker='o')

plt.legend(loc='best')
plt.grid(alpha=0.3)

plt.show()
```

```python
temp = [10, 15, 20, 25, 30, 35, 40]
usage1 = [200, 220, 260, 310, 360, 400, 460]
usage2 = [100, 120, 160, 210, 260, 300, 360]


plt.scatter(temp, usage1, label="usage1")
plt.scatter(temp, usage2, label="usage2")


plt.xlabel(xlabel="Temperature")
plt.ylabel(ylabel="Usages")
plt.title("Electricity use on different temperature.")

plt.legend()
plt.grid(True)
plt.show()

```
