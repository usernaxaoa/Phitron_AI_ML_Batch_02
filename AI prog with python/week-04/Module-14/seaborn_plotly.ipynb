```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Student': ['Ayan', 'Sara', 'John', 'Mira', 'David', 'Nina', 'Rahim', 'Eva', 'Leo', 'Tania', 'Omar', 'Liza'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Age': [18, 19, 20, 21, 22, 18, 19, 20, 21, 22, 23, 24],
    'Math': [72, 78, 75, 85, 88, 74, 80, 79, 87, 90, 93, 95],
    'Science': [68, 74, 77, 82, 84, 70, 76, 78, 83, 86, 89, 91],
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'May', 'May', 'Jun', 'Jun'],
    'City': ['Dhaka', 'Dhaka', 'Chittagong', 'Chittagong', 'Sylhet', 'Sylhet',
             'Dhaka', 'Dhaka', 'Chittagong', 'Chittagong', 'Sylhet', 'Sylhet']
}

df = pd.DataFrame(data)
```

```python
sns.lineplot(data=df, x='Student', y='Math',errorbar=None)
```

```python
sns.lineplot(data=df, x='Student', y='Science',errorbar=None, hue='Gender')
```

```python
sns.relplot(kind='line', data=df, x='Student', y='Math', errorbar=None, hue='Gender')
```

```python
sns.relplot(kind='line', data=df, x='Student', y='Age', errorbar=None, hue='Age')
```
