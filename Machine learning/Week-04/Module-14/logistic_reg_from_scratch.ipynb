```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
x = np.array([
    [10 , 3 ],
    [20 , 5 ],
    [35, 20],
    [50, 10],
])
```

```python
y = np.array([0, 0, 1, 1])
```

```python
def sigmoid(z):
  return 1 / (1+np.exp(-z))
```

```python
w_init=np.array((0.5,-2))
b_init = 1.00
```

```python
def make_prediction(x, w, b, threshold=0.5):
  m = x.shape[0]
  pred_list = np.zeros((m,))

  for i in range(m):
    z = np.dot(w, x[i]) + b
    pred_list[i] = sigmoid(z)

    pred = (pred_list>threshold)
  return pred
make_prediction(x,w_init, b_init)
```
