```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('titanic_data_updated.csv', skipinitialspace=True)
# print(df.shape)
print(f"total potential features: {df.shape[1]}")
print(f"total sample data: {df.shape[0]}")
```

```python
# missing values

df.isnull().sum()
```

```python
# duplicates deleting

print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
sns.countplot(df)
```

```python
data = (df['Survived'].value_counts() / len(df)) * 100
sur_count = df['Survived'].value_counts()
label = df['Survived'].unique()
print(label)
# print(sur_count)
```

```python
plt.pie(data, labels=label, autopct='%1.1F%%')
plt.title("Survival calculation")
plt.legend()
plt.show()
```
