```python
import numpy as np
```

```python
arr = [1,2,3]
print(arr * 5)
```

```python
arr = np.array([1,2,3])
print(arr * 5)
```

```python
arr1 = np.array([1,2,3])
print(arr1.ndim)
print(arr1.dtype)
print(arr1.shape)
```

```python
arr2 = np.array([[1,2,3],[1,2,3]])
print(arr2.ndim)
print(arr2.dtype)
print(arr2.shape)
```

```python
arr3 = np.array([[[1,2], [3,4]], [[1,2],[3,4]]])
print(arr3.ndim)
print(arr3.dtype)
print(arr3.shape)
```

```python
dtypee = np.array([1,2,3.4, '5'], dtype=np.float16)
print(dtypee)
print(dtypee.dtype)
```

```python
dtypee = np.array([1,2,3.4, '5'], dtype=np.float16)
dtypee.astype(np.float16)
print(dtypee)
print(dtypee.dtype)
```

```python
arr = np.array([1,'2', 3.3, True])
print(arr.dtype)
print(arr)
```

```python
lst = [10,20,30,40.7]
arr = np.array(lst)
arr = arr.astype(np.int32)
print(arr)
print(arr.dtype)
print(type(arr))
```

```python
tpl = (1,2,3,4,5)
tpl = np.array(tpl)
print(type(tpl))
print(tpl.dtype)
print(tpl.ndim)
print(tpl.shape)
```

```python
st = {1,2,3,4}
st = np.array((st))
print(type(st))
print(st.dtype)
print(st.ndim)
print(st.shape)
```

```python
dc = {'a': 10, 'b': 20, 'c': 30}

key = dc.keys()
value = dc.values()
item = dc.items()

k = np.array(key)
print(type(k))
print(k)

v = np.array(value)
print(type(v))
print(v)

i = np.array(list(item))
print(type(i))
print(i)
```

```python
arr = np.random.rand(3,3) 
print(arr)
print(arr.dtype)
```

```python
arr = np.random.randint(1, 10, (10,10))
print(arr)
print(arr.dtype)
```

```python
arr = np.random.uniform(1, 10, (3,3))
print(arr)
print(arr.dtype)
```

```python
arr = np.arange(1,10, 3).reshape(3,)
print(arr)
```

```python
arr = np.linspace(1, 10, 19)
print(arr)
```

```python
arr = np.logspace(0, 4, 5)
print(arr)
```
