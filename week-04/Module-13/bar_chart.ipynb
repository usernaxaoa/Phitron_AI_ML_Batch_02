```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
category = ['a', 'b', 'c', 'd', 'e']
values = [10, 20, 25, 15, 17]

plt.bar(category, values, color='blue', edgecolor='black')
```

```python
category = ['a', 'b', 'c', 'd', 'e']
values = [10, 20, 25, 15, 17]

plt.barh(category, values, color='blue', edgecolor='black')
```

```python
category = ['a', 'b', 'c', 'd', 'e']

x = np.arange(len(category))
value1 = [25,40,30,20,35]
value2 = [30,35,25,28,40]

plt.bar(x+0.1, value1, width=0.2, color='black', label='2024')
plt.bar(x-0.1, value2, width=0.2, color='blue', label='2025')
plt.xticks(x, category)
plt.legend()
plt.show()

```

```python
# practice question

brands = ['brand a', 'brand b', 'brand c']
sales_24 = [250, 400, 350]
sales_25 = [300, 420, 370]
```

```python
# vertical bar 
x = np.arange(len(brands))
plt.bar(x+0.2, sales_24, width=0.4, color='red', edgecolor='white', label=2024)
plt.bar(x-0.2, sales_25, width=0.4, color='green', edgecolor='white', label=2025)
plt.xticks(x,brands)
plt.legend()
plt.show()
```

```python
# horizontal bar 
x = np.arange(len(brands))
plt.barh(x+0.2, sales_24, height=0.4, color='red', edgecolor='white', label=2024)
plt.barh(x-0.2, sales_25, height=0.4, color='green', edgecolor='white', label=2025)
plt.yticks(x,brands)
plt.legend()
plt.show()
```
