```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = np.random.normal(50, 10, 200)
```

```python
plt.style.use('seaborn-v0_8-dark')
plt.hist(data, bins=5, color='red', edgecolor='white')
plt.title('Basic Histogram')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
np.random.seed(0)
rand_exam_score = np.random.randint(0, 100, 300)

plt.hist(rand_exam_score, bins=50, color='black', edgecolor='white')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Exam score frequency")
plt.grid(axis='y', color="black")
plt.show()
```
