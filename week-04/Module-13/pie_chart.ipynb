```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
label = ['a', 'b', 'c', 'd']
value = [30, 25, 20, 25]

plt.pie(value, labels=label)
plt.show()
```

```python
colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']
explode = [0.05, 0, 0, 0]

plt.pie(value, labels=label, colors=colors, autopct='%1.1f%%', startangle=90, explode=explode, shadow=True)
plt.title('Customized Pie Chart')
plt.legend()
plt.show()
```

```python
plt.pie(value, labels=label, colors=colors, autopct='%1.1f%%')
plt.legend(title='Categories')
plt.title('Pie Chart with Legend')
plt.show()
```

```python
products = ['pa', 'pb', 'pc', 'pd']
share = [40,25,20,15]

plt.pie(share, labels=products, startangle=180, autopct='%1.1f%%', shadow=1, explode=explode)
plt.legend()
plt.show()
```
